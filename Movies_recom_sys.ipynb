```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D plotting
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import TfidfVectorizer
from rapidfuzz import process, fuzz

# Load your merged dataset
file_path = 'combined_dataset.csv'  # <-- Your dataset path
df = pd.read_csv(file_path)

# Corrections for messy platform names
platform_corrections = {
    'Prime Video': 'Amazon Prime Video',
    'Amazon Prime': 'Amazon Prime Video',
    'Amazon Video': 'Amazon Prime Video',
    'Amazon Prime(coming on25/04/25)': 'Amazon Prime Video',
    'Netflix/Amazon Prime': 'Netflix, Amazon Prime Video',
    'Netflix Kids': 'Netflix',
    'MUBI Amazon Channel': 'MUBI',
    'ManoramaMAX Amazon Channel': 'ManoramaMAX',
    'ManoramaMax Amazon Channel': 'ManoramaMAX',
    'Mubi': 'MUBI',
    'mubi': 'MUBI',
    'Chaupal Amazon Channel': 'Chaupal',
    'Chaupal Amazon channel': 'Chaupal',
    'Chaupal prime video channels': 'Chaupal',
    'Crunchyroll Amazon Channel': 'Crunchyroll',
    'Crunchyroll Amazon Channel or for free with ads on Crunchyroll': 'Crunchyroll',
    'Crunchyroll Amazon Channel or for free with ads on Crunchyroll.': 'Crunchyroll',
    'Discovery+ Amazon Channel': 'Discovery+',
    'Anime Times Amazon Channel': 'Anime Times',
    'Lions Gate Play/Apple Tv': 'Lionsgate Play, Apple TV',
    'Lionsgate Play Apple Tv Channel': 'Lionsgate Play, Apple TV',
    'Lionsgate Play Amazon channel': 'Lionsgate Play',
    'Free on MX Player': 'MX Player',
    'Google play': 'Google Play',
    'Google Play': 'Google Play',
    'Apple Tv': 'Apple TV',
    'AppleTV': 'Apple TV',
    'Apple TV+': 'Apple TV',
    'GuideDoc': 'GuideDoc',
    'WOW Presents Plus': 'WOW+',
    'Sun NXT': 'Sun NXT',
    'Sun Nxt': 'Sun NXT',
    'SonyLiv': 'Sony Liv',
    'YouTube Paid': 'YouTube',
    'Youtube': 'YouTube',
    'aha': 'Aha',
    'plex': 'Plex',
    'Just Watch/Plex': 'Just Watch, Plex',
    'Disney+ Hotstar': 'Hotstar',
    'Amazon Prime/JioHotstar': 'Amazon Prime Video, Hotstar',
    'Rent / Buy on Apple Tv': 'Rent it on Apple TV',
    'Rent it on Apple TV': 'Rent it on Apple TV',
    'Rent/Buy on Apple TV': 'Rent it on Apple TV',
    'Rent/Buy on Apple Tv': 'Rent it on Apple TV',
    'Rent on Apple TV': 'Rent it on Apple TV',
    'Rent on Amazon Video': 'Amazon Prime Video',
    'Rent it on Google Play Movies': 'Google Play'
}

# Clean and split platform names
def clean_and_split(platform):
    if isinstance(platform, float):
        return []
    parts = [p.strip() for p in platform.split(',') if p.strip()]
    corrected_parts = [platform_corrections.get(p, p) for p in parts]
    return corrected_parts

# Apply cleaning
df['Platform'] = df['Platform'].apply(clean_and_split)
df['Alternatives'] = df['Alternatives'].fillna('0').apply(lambda x: clean_and_split(x) if x != '0' else [])

# Build final platform list
all_platforms = set()
for platforms in df['Platform']:
    all_platforms.update(platforms)
for alts in df['Alternatives']:
    all_platforms.update(alts)

unwanted_platforms = {
    'Not Available For Streaming', 'Not available for streaming', 'Not Available',
    'not available for streaming', '0', 'Theatre', 'BookMyShow', 'No Information',
    'No information', 'JIoHotstar', 'Jio Hotstar', 'JioHotstar',
    'amazon video-(leaving 24th april)', 'bookmyshow (Cinema)'
}

final_platforms = sorted({plat for plat in all_platforms if plat not in unwanted_platforms})
final_platforms = sorted(set(p.title() for p in final_platforms))

# -------- Program Starts --------

print("Welcome to the Movie/Series Finder!\n")

selected_ott = ['Netflix', 'Amazon Prime Video', 'Hotstar', 'Sony Liv',
                'Apple TV', 'Zee5', 'MX Player', 'Jio Cinema', 'YouTube', 'Aha']

search_specific = input("Are you looking for a specific movie/series? (yes/no): ").lower()

if search_specific == 'yes':
    query = input("Enter the name of the movie/series: ").strip()
    titles_list = df['Title'].dropna().tolist()

    matches = process.extract(query, titles_list, limit=5, scorer=fuzz.WRatio)
    best_matches = [match[0] for match in matches if match[1] > 60]

    if best_matches:
        print("\nTop Matches:")
        for title in best_matches:
            result = df[df['Title'] == title].iloc[0]
            print(f"\nTitle: {result['Title']}")
            print(f"Year Released: {result['Year Released']}")
            print(f"Seasons: {result['Seasons']}")
            print(f"Platform: {', '.join(result['Platform'])}")
            print(f"Alternatives: {', '.join(result['Alternatives'])}")
            print(f"Type: {result['Movie/Web-series']}")

            platforms_list = result['Platform'] + result['Alternatives']
            available_on_other_platforms = [plat for plat in platforms_list if plat not in selected_ott]
            print(f"Available on other platforms: {', '.join(available_on_other_platforms) if available_on_other_platforms else 'Not available on other platforms.'}")

        # Visualization Part 1: Nearest Neighbors (3D)
        all_titles = df['Title'].dropna().tolist()
        similarities = [fuzz.WRatio(query, title) for title in all_titles]
        sim_df = pd.DataFrame({'Title': all_titles, 'Similarity': similarities})
        top_neighbors = sim_df.sort_values(by='Similarity', ascending=False).head(6)
        vectorizer = TfidfVectorizer()
        vectors = vectorizer.fit_transform(top_neighbors['Title'])
        pca = PCA(n_components=3)
        points = pca.fit_transform(vectors.toarray())

        fig = plt.figure(figsize=(10,8))
        ax = fig.add_subplot(111, projection='3d')
        for idx, (x, y, z) in enumerate(points):
            title = top_neighbors['Title'].iloc[idx]
            if title.lower() == query.lower():
                ax.scatter(x, y, z, color='blue', s=120, label='Input Movie' if idx==0 else "")
                ax.text(x+0.01, y+0.01, z+0.01, title, fontsize=9, color='blue')
            else:
                ax.scatter(x, y, z, color='green', s=70, label='Recommended' if idx==1 else "")
                ax.text(x+0.01, y+0.01, z+0.01, title, fontsize=8, color='green')
        ax.set_title('Nearest Neighbor Movies (based on Title Similarity)')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        handles, labels = ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        ax.legend(by_label.values(), by_label.keys())
        plt.show()

        # === ✅ Visualization Part 2: Available Platforms (3D Bar Plot - Updated) ===
        result = df[df['Title'].str.lower() == best_matches[0].lower()].iloc[0]
        platforms_list = result['Platform'] + result['Alternatives']

        if platforms_list:
            fig = plt.figure(figsize=(12,8))
            ax = fig.add_subplot(111, projection='3d')

            xpos = np.arange(len(platforms_list))
            ypos = np.zeros_like(xpos)
            zpos = np.zeros_like(xpos)

            dx = dy = np.ones_like(xpos)
            dz = np.ones_like(xpos)

            colors = plt.cm.tab20(np.linspace(0,1,len(platforms_list)))

            ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color=colors, shade=True)

            ax.set_yticks([])
            ax.set_zlim(0, 1.5)
            ax.set_xticks(xpos + 0.5)  # Shift labels to center
            ax.set_xticklabels(platforms_list, rotation=30, ha='right', fontsize=10)
            ax.set_title('Available Platforms for this Movie/Series')
            ax.set_xlabel('Platforms')
            plt.tight_layout()
            plt.show()
        else:
            print("\nNo alternative platforms found for this movie.")

    else:
        print("No close matches found.")

else:
    print("\nAvailable Platforms:")
    available_platforms = set()
    for idx, row in df.iterrows():
        for platform in row['Platform'] + row['Alternatives']:
            if platform in selected_ott:
                available_platforms.add(platform)

    available_platforms = sorted(available_platforms)

    for idx, platform in enumerate(available_platforms, 1):
        print(f"{idx}. {platform}")

    user_input = input("\nEnter your preferred platform(s), separated by commas: ").split(',')
    selected_platforms = [p.strip() for p in user_input]

    def is_available(row):
        return any(p in row['Platform'] for p in selected_platforms) or any(p in row['Alternatives'] for p in selected_platforms)

    found_titles = df[df.apply(is_available, axis=1)]

    if not found_titles.empty:
        print("\nTitles available on your selected platform(s):\n")
        for idx, row in found_titles.iterrows():
            print(f"Title: {row['Title']} | Type: {row['Movie/Web-series']} | Platform(s): {', '.join(row['Platform'])}")

            platforms_list = row['Platform'] + row['Alternatives']
            available_on_other_platforms = [plat for plat in platforms_list if plat not in selected_platforms]
            print(f"Available on other platforms: {', '.join(available_on_other_platforms) if available_on_other_platforms else 'Not available on other platforms.'}")

    else:
        print("No titles found for the selected platform(s).")
```
